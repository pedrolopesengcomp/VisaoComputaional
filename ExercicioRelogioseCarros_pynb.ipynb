Importando CV2

```python
import cv2
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
```

```python
relogios = cv2.imread('/content/drive/MyDrive/VisaoComputacionaç/Images/clock.jpg')
carros = cv2.imread('/content/drive/MyDrive/VisaoComputacionaç/Images/car.jpg')

cv2_imshow(relogios)
cv2_imshow(carros)
```

Processando imagens

```python
carrosBlack = cv2.cvtColor(carros, cv2.COLOR_BGR2GRAY)
relogiosBlack = cv2.cvtColor(relogios, cv2.COLOR_BGR2GRAY)
```

Detectando objetos

```python
relogios = cv2.imread('/content/drive/MyDrive/VisaoComputacionaç/Images/clock.jpg')
carros = cv2.imread('/content/drive/MyDrive/VisaoComputacionaç/Images/car.jpg')

detector_carros = cv2.CascadeClassifier('/content/drive/MyDrive/VisaoComputacionaç/Cascades/cars.xml')
detector_relogios = cv2.CascadeClassifier('/content/drive/MyDrive/VisaoComputacionaç/Cascades/clocks.xml')

deteccoesCarros = detector_carros.detectMultiScale(carrosBlack,
                                                   scaleFactor=1.1,
                                                   minNeighbors=1,
                                                   maxSize=(60,60),
                                                   minSize=(24,24))

deteccoesRelogios = detector_relogios.detectMultiScale(relogiosBlack,
                                                       scaleFactor=1.03,
                                                       minNeighbors=1)

print(deteccoesRelogios)

for x,y,w,h in deteccoesCarros:
  cv2.rectangle(carros, (x,y), (x+w, y+h), (0,255,0), 2)

for x,y,w,h in deteccoesRelogios:
  cv2.rectangle(relogios, (x,y), (x+w, y+h), (0,255,0), 2)

cv2_imshow(relogios)
```
